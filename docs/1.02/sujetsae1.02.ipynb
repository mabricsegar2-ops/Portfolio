```python
liste = []

monset = set()
with open('donnees_source.txt','r', encoding = 'utf8') as f:
    for i in range(7):
        ligne = f.readline()
        if '=' in ligne:
            ligne = ligne.split('=')
            liste.append(ligne[0])
    ligne = f.readline()
    while(ligne!=""):
        if('#Code produit') in ligne:
            ligne = ligne.split('=')
            monset.add(ligne[1].upper())
        ligne = f.readline()

print(liste)
print(monset)
for i in liste :
    print(i)
for i in monset:
    print(i[:-1])

with open('texte.txt','w', encoding = 'utf8') as f:
    for i in liste :
        f.write(i+";")
    for i in monset:
        f.write(i[:-1]+";")
    f.write("PrixTotal\n")
```
